```python
import pickle, numpy as np

def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict

def load_cifar100(path="/kaggle/input/cifar-100"):
    train = unpickle(f"{path}/train")
    test = unpickle(f"{path}/test")

    X_train = train[b'data'].reshape(-1, 3, 32, 32).transpose(0,2,3,1)
    y_train = np.array(train[b'fine_labels'])
    X_test = test[b'data'].reshape(-1, 3, 32, 32).transpose(0,2,3,1)
    y_test = np.array(test[b'fine_labels'])
    return X_train, y_train, X_test, y_test

X_train, y_train, X_test, y_test = load_cifar100()
print(X_train.shape, y_train.shape)  # (50000, 32, 32, 3) (50000,)
```

```python
import pickle

def load_meta(path="/kaggle/input/cifar-100/meta"):
    with open(path, 'rb') as fo:
        meta = pickle.load(fo, encoding='bytes')
    fine_names = [n.decode('utf-8') for n in meta[b'fine_label_names']]
    coarse_names = [n.decode('utf-8') for n in meta[b'coarse_label_names']]
    return fine_names, coarse_names

fine_names, coarse_names = load_meta()
print("Fine classes:", fine_names[:10])    # first 10 fine labels
print("Coarse classes:", coarse_names)     # 20 superclasses
```
