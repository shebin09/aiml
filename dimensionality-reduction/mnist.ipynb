```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
data1=pd.read_csv('mnist_train1.csv')
```

```python
X=data1.drop(columns=['label'])
```

```python
pca=PCA()
mn_new=pca.fit_transform(X)
explained_variance=pca.explained_variance_ratio_
explained_variance
np.cumsum(explained_variance)
```

```python
import matplotlib.pyplot as plt
plt.bar(range(1,len(explained_variance)+1),explained_variance)
plt.xlabel('principal components')
plt.ylabel('explained variance')
plt.show()
```

```python
pca=PCA(n_components=2)
mn_new_2=pca.fit_transform(X)
```

```python
exp_new=pca.explained_variance_ratio_
exp_new
np.sum(exp_new)
```

```python
y=data1['label']
plt.figure(figsize=(10, 8))
plt.scatter(mn_new_2[:, 0], mn_new_2[:, 1], c=y, cmap='tab10', s=1)
plt.colorbar()
plt.title("MNIST visualized in 2D via PCA")
plt.show()
```
