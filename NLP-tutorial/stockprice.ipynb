```python
import pandas as pd
```

```python
df=pd.read_csv('Data.csv',encoding='ISO-8859-1')
```

```python
df
```

```python
train=df[df['Date']< '20150101']
test=df[df['Date']>'20141231']
```

```python
data=train.iloc[:,2:27]
data.replace('[^a-zA-Z]'," ",regex=True,inplace=True)
```

```python
data
```

```python
new_index=[str(i) for i in range(25)]
new_index
data.columns=new_index
```

```python
for i in new_index:
    data[i]=data[i].str.lower()
```

```python
headlines=[]

for row in range(0,len(data.index)):
    headlines.append(" ".join(str(x) for x in data.iloc[row,0:25]))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
countvec=CountVectorizer(ngram_range=(2,2))
traindataset=countvec.fit_transform(headlines)
```

```python
model=RandomForestClassifier(n_estimators=200,criterion='entropy')
model.fit(traindataset,train['Label'])
```

```python
test_transform=[]
for row in range(0,len(test.index)):
    text=' '.join(str(x) for x in test.iloc[row,2:27])
    test_transform.append(text)
test_data=countvec.transform(test_transform)
pred=model.predict(test_data)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score

acc=accuracy_score(pred,test['Label'])
acc
```

```python
import joblib

joblib.dump(model, "model.pkl")
joblib.dump(countvec, "countvec.pkl")
```
