```python
import matplotlib.pyplot as plt
x=[0,1,2,3,4,5]
y=[0,3,2,22,22,10]
plt.plot(x,y)
plt.xlabel("days")
plt.ylabel("counts")
plt.xlim(1,5)
plt.ylim(1,10)

plt.show()
```

```python
import numpy as np

f=50
x=np.arange(0,400)
y=np.cos(2*np.pi*f*x/2000)
plt.plot(x,y)
plt.xlim(300,400)
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
```

```python
x=[0,1,2,3,4,5]
y1=[0,3,2,22,22,10]
y2=[1,7,5,4,3,8]
y3=[2,3,4,5,5,6]
plt.plot(x,y1,c='g',label='location1')
plt.plot(x,y2,c='r',label='location2')
plt.plot(x,y3,c='b',label='location3')
plt.xlabel('days')
plt.ylabel('counts')
plt.legend()
plt.show()
```

```python
x=['monday','tuesday','wednesday','thurs','friday']
item1=[1,2,3,4,5]
plt.bar(x,item1,label='item1')
plt.xlabel('days')
plt.ylabel('count')
plt.savefig('bar_chart.png',dpi=300)
plt.legend()
plt.show()
```

```python
x=['monday','tuesday','wednesday','thurs','friday']
item1=[1,2,3,4,5]
item2=[2,4,1,6,5]
plt.bar(x,item1,color='r', label='item1')
plt.bar(x,item2,color='b' ,alpha=0.2,label='item2')
plt.xlabel('days')
plt.ylabel('count')
plt.savefig('bar_chart1.png',dpi=300)
plt.legend()
plt.show()
```

```python
countries=['india','USA','UK','UAE']
popuplation=[123,20,30,40]
plt.pie(popuplation,labels=countries,autopct='%1.2f%%')
plt.savefig('pie_chart.png',dpi=300)
plt.show()
```

```python
import pandas as pd
data=pd.read_csv('auto-mpg.csv')
plt.hist(data['mpg'],color='green',rwidth=0.9,density=True)


plt.title("hist of mpg")
plt.xlabel('MPG')
plt.ylabel("Density")
plt.show()

```

```python

data['mpg'].plot.kde()
plt.show()
```

```python
import folium

latitude=8.5559403
longitude=76.87630
my_map= folium.Map(location=[latitude,longitude],zoom_start=50)
folium.Marker([latitude,longitude],popup="technopark").add_to(my_map)
my_map.save("tvm.html")
```
