```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
heat_data = np.array([[0,0.4,1],[1,0.4,0]])
sns.heatmap(heat_data,annot=True)
```

```python
data=pd.read_csv('auto-mpg.csv')
data_new=data.drop(['car name','horsepower'],axis=1)
corr_mat=data_new.corr()
mask=np.triu(np.ones_like(corr_mat,dtype=bool))

sns.heatmap(corr_mat,mask=mask,annot=True)
plt.xticks(rotation=45)
plt.show()
```

```python
num_df=data.select_dtypes(include=np.number)
plt.boxplot(num_df)
plt.show()
```

```python
import numpy as np
fig1,ax1=plt.subplots()
f=50
x=np.arange(0,400)
y1=np.cos(2*np.pi*f*x/2000)
y2=np.sin(2*np.pi*f*x/2000)
ax2=ax1.twinx()
ax1.plot(x,y1,c='k',label='$y=cos(x)$')
ax2.plot(x,y2,c='b',label='$y=sin(x)$')
plt.legend()
plt.xlabel("time")
plt.ylabel("amplitude")
plt.text(100,1,"some waves")
plt.grid()
plt.show()
```

```python
x=np.arange(0,400).reshape(20,20)
plt.imshow(x)
plt.yticks([])
plt.show()
```

```python
plt.figure(figsize=(15,12))
plt.subplot(2,2,1)
a=data['model year'].value_counts().sort_index()
plt.bar(a.index,a.values)
plt.xlabel('number of cars')
plt.ylabel('year')
plt.title('year vs no. of cars')
plt.subplot(2,2,2)
plt.scatter(data['weight'],data['acceleration'])
plt.xlabel('weight')
plt.ylabel('acceleration')
plt.title('weight vs acceleration')
plt.subplot(2,2,3)
plt.boxplot(num_df)
plt.title('boxplot')
plt.subplot(2,2,4)
sns.heatmap(num_df.corr(),annot=True)
plt.title('heatmap of correlations')
plt.yticks(rotation=45)
plt.savefig('dashboard.png')
plt.show()
```
