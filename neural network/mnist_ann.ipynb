```python
import pandas as pd
```

```python
data=pd.read_csv('mnist_train1.csv')
```

```python
data
```

```python
X=data.drop(columns=['label'])
y=data['label']
X.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input,Dropout,Dense
from tensorflow.keras.regularizers import L1,L2,L1L2
```

```python
model =Sequential()
model.add(Input(shape=(784,)))
model.add(Dense(512,activation='relu',kernel_initializer='he_uniform',kernel_regularizer=L1(l1=1e-3)))
model.add(Dense(256,activation='relu',kernel_initializer='he_uniform',kernel_regularizer=L1(l1=1e-3)))

model.add(Dense(128,activation='relu',kernel_initializer='he_uniform',kernel_regularizer=L1(l1=1e-3)))
model.add(Dense(10,activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
model_hist=model.fit(X_train,y_train,validation_data=(X_test,y_test),batch_size=64,epochs=30)
```

```python
model.get_metrics_result()
```

```python
model1 =Sequential()
model1.add(Input(shape=(784,)))
model1.add(Dense(512,activation='relu',kernel_initializer='he_uniform',kernel_regularizer=L2(l2=1e-3)))
model1.add(Dense(256,activation='relu',kernel_initializer='he_uniform',kernel_regularizer=L2(l2=1e-3)))
model1.add(Dense(128,activation='relu',kernel_initializer='he_uniform',kernel_regularizer=L2(l2=1e-3)))
model1.add(Dense(10,activation='softmax'))
```

```python
model1.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model_hist1=model1.fit(X_train,y_train,validation_data=(X_test,y_test),batch_size=64,epochs=30)
```

```python
model2 =Sequential()
model2.add(Input(shape=(784,)))
model2.add(Dense(512,activation='relu',kernel_initializer='he_uniform'))
model2.add(Dropout(0.2))
model2.add(Dense(256,activation='relu',kernel_initializer='he_uniform'))
model2.add(Dropout(0.2))
model2.add(Dense(128,activation='relu',kernel_initializer='he_uniform'))
model2.add(Dropout(0.2))
model2.add(Dense(10,activation='softmax'))
```
